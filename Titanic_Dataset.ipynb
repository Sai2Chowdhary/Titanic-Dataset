```python
import pandas as pd
df=pd.read_csv('train.csv')
df.head()
```

```python
#Data Exploriation
df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis='columns',inplace=True)
df.head()
```

```python
target=df.Survived	
inputs=df.drop('Survived',axis='columns')
inputs.head()
```

```python
#Here we have sex column with text column as we know that machine learning model can't handle text
#so, convert it into dummies(numbers)
dummies=pd.get_dummies(inputs.Sex)
dummies.head()
```

```python
#concating the two columns inputs and dummies
inputs=pd.concat([inputs,dummies],axis='columns')
inputs.head()
```

```python
#drop sex column because we have dummy columns
inputs.drop('Sex',axis='columns',inplace=True)
inputs.head()

```

```python
#isna() is used to detect missing values.
#return a boolean same sized object indicating if the values are NA.
inputs.columns[inputs.isna().any()]
```

```python
#lets see the example of NaN
inputs.Age[:10]
```

```python
#Here we have NaN values 
#we handle these values by taking mean 
#fillna() is used to fill the missing values 
inputs.Age=inputs.Age.fillna(inputs.Age.mean())
inputs.head()
```

```python
#Now we can see that missing values are filled
inputs.Age[:10]
```

```python
#We use sklearn model to seperate training and testing data
#we divide in 80:20 ratio
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(inputs,target,test_size=0.2) 
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
#we can see our train data
X_train
```

```python
#Creating Naive Bayes model using Gaussian naive bayes model
from sklearn.naive_bayes import GaussianNB
model=GaussianNB()
```

```python
#training the model
#we have call the fit method when ever we want to train the model
model.fit(X_train,Y_train)
```

```python
#measuring the accuracy
model.score(X_test,Y_test)
```
